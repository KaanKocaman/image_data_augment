## OpenCV Kütüphanesi

```python
import cv2
import matplotlib.pyplot as plt
import os
import glob

# Klasör yolunu tanımladık
folder_path = r'C:\Users\meteh\archive (3)\train\urban'  

# Klasördeki tüm .jpg dosyalarını listeledik
image_paths = glob.glob(os.path.join(folder_path, '*.jpg'))

# Görüntü sayısını aldık
num_images = len(image_paths)

# Görüntüleri yan yana göstermek için bir figür oluştur
plt.figure(figsize=(10, 10)) 

# Görüntüleri işlemek ve göstermek için döngü
for i, image_path in enumerate(image_paths):
    # Görüntüyü okuma işlemini yaptık
    img = cv2.imread(image_path)
    
    # BGR formatını RGB formatına çevirdik
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Yan yana göstermek için subplot oluşturduk
    plt.subplot(1, num_images, i + 1)
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title(os.path.basename(image_path))  # Görüntü dosya adını başlık olarak ekledik

plt.tight_layout()  
```

```python
import cv2
import matplotlib.pyplot as plt
import os
import glob

folder_path = r'C:\Users\meteh\archive (3)\train\urban'

file_name="urban_44.jpg"
file_path=os.path.join(folder_path,file_name)

img=cv2.imread(file_path)

# BGR formatını RGB formatına çevirdik
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Gri tonlama ekledik
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Gri tonlamalı görüntüyü aynı klasör içine kaydettik
gray_file_name="urban_44_gray.jpg"
gray_file_path=os.path.join(folder_path,gray_file_name)
cv2.imwrite(gray_file_path,gray)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Orijinal Görsel")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(gray,cmap="gray")
plt.title("Gri Tonlanmış Görsel")
plt.axis("off")

plt.show()
print(f"Gri tonlamalı görüntü kaydedildi: {gray_file_path}") 
#Gri tonlamalı görüntünün aynı klasör içine kaydedildiği bilgisini verdik
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Klasör yolunu tanımla
folder_path = r'C:\Users\meteh\archive (3)\train\urban'

# Klasördeki tüm .jpeg dosyalarını listele
image_paths = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.jpeg')]

# Her görüntü için işlemi yapmak için for döngüsüm kurduk
for image_path in image_paths:
    img = cv2.imread(image_path)

    # Keskinleştirme filtresi uyguladık
    blur = cv2.GaussianBlur(img, (0, 0), 3)
    sharpened = cv2.addWeighted(img, 1.5, blur, -0.5, 0)

    # Görüntüleri RGB formatına çevir
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    sharpened_rgb = cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB)

    # Orijinal ve keskinleştirilmiş görüntüleri yan yana gösterdik
    plt.figure(figsize=(12, 6))
    
    # Orijinal görüntü
    plt.subplot(1, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Orijinal Görüntü")
    plt.axis('off')
    
    # Keskinleştirilmiş görüntü
    plt.subplot(1, 2, 2)
    plt.imshow(sharpened_rgb)
    plt.title("Keskinleştirilmiş Görüntü")
    plt.axis('off')

    plt.show()
    break  # Sadece ilk görüntüyü göstermek için döngüden çık
```

## skimage kütüphanesi

```python
from skimage import io, transform, color
import matplotlib.pyplot as plt
import os

# Klasör yolunu tanımla
folder_path = r'C:\Users\meteh\archive (3)\train\urban'

# Belirli dosyanın yolunu oluştur
file_name = 'urban_45.jpg'
file_path = os.path.join(folder_path, file_name)

# Görüntüyü oku
img = io.imread(file_path)

# Döndürme açıları
angles = [0, 45, 90, 135, 180]

# Döndürülmüş görüntüleri saklamak için liste
rotated_images = []

# Görüntüyü belirtilen açılarla döndür
for angle in angles:
    rotated_img = transform.rotate(img, angle)
    rotated_images.append(rotated_img)

# Görüntüleri yan yana göster
plt.figure(figsize=(20, 6))

for i, rotated_img in enumerate(rotated_images):
    plt.subplot(1, 5, i + 1)
    plt.imshow(rotated_img) 
    plt.title(f'{angles[i]}°')
    plt.axis('off')

plt.show()
```

## PIL kütüphanesi


```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
import os

# Klasör yolunu tanımla
folder_path = r'C:\Users\meteh\archive (3)\train\urban'

# Belirli dosyanın yolunu oluştur
file_name = 'urban_37.jpg'
file_path = os.path.join(folder_path, file_name)

# Görüntüyü oku
img = io.imread(file_path)

# NumPy dizisini PIL Image nesnesine dönüştür
img_pil = Image.fromarray(np.uint8(img)) 

# Yatay ve dikey yansıtma işlemleri
horizontal = img_pil.transpose(Image.FLIP_LEFT_RIGHT)
vertical = img_pil.transpose(Image.FLIP_TOP_BOTTOM)

# Görüntüleri göster
plt.figure(figsize=(15, 5))

# Orijinal görüntü
plt.subplot(1, 3, 1)
plt.imshow(img_pil)
plt.title("Orijinal")
plt.axis("off")

# Yatay yansıtılmış görüntü
plt.subplot(1, 3, 2)
plt.imshow(horizontal)
plt.title("Yatay")
plt.axis("off")

# Dikey yansıtılmış görüntü
plt.subplot(1, 3, 3)
plt.imshow(vertical)
plt.title("Dikey")
plt.axis("off")

plt.show()
```
